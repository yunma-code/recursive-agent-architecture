```python
# Define multiple agents (prompt sets)
prompt_sets = {
    "metadata": {
        "system_prompt": "You are a metadata extractor for technical documents. Be concise.",
        "prompts": {
            "document_title": "What is the title or main topic of the document? Respond in <1 sentence.",
            "document_type": "What type of document is this? Respond in <3 words.",
            "document_purpose": "What is the purpose of this document? Respond in <1 sentence.",
            "document_summary": "Summarize the document in 2-3 sentences.",
            "content_type": "What is the type of content contained in the document? Respond in <3 words.",
            "content_structure": (
                  "List only the top-level sections of the document. "
                  "Number them (e.g., 1., 2., 3.) and do not include subpoints or bullet points."
                  "Provide at least 3-5 sections."
            )
        }
    },
    "analyze_document": {
        "system_prompt": "You are a technical document analyst deciding if you have enough information for a patent.",
        "prompts": {
            "contains_technical_problem":"TRUE/FALSE: This document describes, references, or infers a technical problem. Explain your reasoning in <1 sentence.",
            "contains_technical_solution":"TRUE/FALSE: This document describes, references, or infers a technical solution. Explain your reasoning in <1 sentence.",
            "provides_technical_implementation":"TRUE/FALSE: This document provides a specific technical implementation of the technical solution. Explain your reasoning in <1 sentence.",
            "easily_replicable":"TRUE/FALSE: A person skilled in the domain could replicate the technical solution based solely on the document. Explain your reasoning in <1 sentence.",
            "solution_is_enabled":"TRUE/FALSE: The document provides enough specific technical implementation detail that the solution is \"enabled\". Explain your reasoning in <1 sentence.",
            "contains_enough_detail_for_patent_attorney":"TRUE/FALSE: This document contains enough detail for a patent attorney to immediately draft a patent application. Explain your reasoning in <1 sentence.",
            "sub-prompts": {
                 "missing_info": "What else would a patent attorney need to write-up a patent application that's missing in the document?"
            }
        }
    },
    "validate_document_depth":{
        "system_prompt": "You are a technical document analyst deciding if you have enough information for a patent.",
        "prompts": {
            "technical_maturity": "rate the technical maturity of the solution mentioned and explain your reasoning.",
            "technical_difficulty": "rate the technical difficulty of implementing the solution mentioned and explain your reasoning.",
            "potential_areas_of_novelty": "identify possible areas of opportunity for patenting.",
            "sub-prompts":{
                "recommendation": "Decide whether we have enough information to proceed or if we need more information.",
                "missing_info": "identify specific details that are missing that you'd need to determine if you can move forward.",
                "questions": "List up to three simple questions you can pose to the user that will give you the info you need to move forward with a patent application."
            }
        }
    },
    "technical_characteristics": {
        "system_prompt": "You are a metadata extractor for technical documents. Be concise.",
        "prompts": {
            "industry_domain": "List up to 5 industry domains associated with the document.",
            "technical_domain": "List up to 5 technical domains associated with the document.",
            "sub-prompts": {
                "technical_concepts": "For each technical domain, list up to 5 techniques, concepts, functions, and/or components used.",
            }
        }
    },
    "core_problem_extractor": {
        "system_prompt": "You are a distinguished engineer with deep technical aptitude and expert problem-scoping skills",
        "prompts": {
            "core_problem":"""Describe the core technical problem or unmet user need in a 5-sentence paragraph. Use specific examples to showcase how it is currently experienced. Focus on conveying well-scoped engineering problems that consider specific fundamental constraints that are innate to the domain. Be direct, specific, and tangible. Ensure that each of your responses scores highly on this rubric: Clarity: the problem is stated so that both technical and non-technical stakeholders can understand. Relevance: the problem is rooted in a specific real-world need. Specificity: the problem describes how a specific stakeholder directly experiences the core problem or unmet need. Scope: the problem is specific about the extent of the problem, including use cases, markets, or industries affected.""",
            "decomposed_problem": "Deconstruct the problem into an exhaustive and complete list of sub-problems and problem components so that you fully understand it.",
            "causal_issues":"For every sub-problem or problem component, use the 5 \"Why's\" framework on each to develop a deep and nuanced understanding of fundamental causal elements.",
            "specific_root_causes": "Identify the core root causes of the problem to understand what's truly important. Differentiate between symptoms vs. root causes. Root Causes are the underlying, fundamental issues that contribute to the existence of the core problem. They represent the challenges that the invention seeks to address and, if mitigated or resolved, would directly impact the effectiveness of the invention. Identifying the root causes will inform the technical challenges and ultimately shape the scope of the patent claims. Characteristics of strong root causes include: Technical Limitations: must be technical, focusing on how something works and what limitations of technology prevent a solution, rather than what it does. Causal: should demonstrate a clear cause-and-effect relationship, explaining why existing solutions are inadequate and how they contribute to the problem. Fundamental: should be the deepest, most fundamental problem, not merely a symptom of another issue.",
            "subprompts": {
                "design_constraints": "Using your analysis above, generate an exhaustive list of the fundamental technical limitations that inhibit a solution and specific real-world design considerations you must consider in a solution.",
                "technical_constraints" : "Generative an exhaustive and complete list of specific technical limitations you expect to encounter head-on in implementing a solution.",
                "tradeoffs": "Apply the TRIZ framework exhaustively to frame each specific causal issue as a fundamental technical constraint or tradeoff."
            }
       }
    },
    "use_case_analysis": {
        "system_prompt": "You are a strategic analyst identifying technical use cases.",
        "prompts": {
            "problem_solved": "What problem is this system solving?",
            "end_users": "Who are the end users or beneficiaries of this system?"
        }
    }
}

# Core runner that supports sub-prompts
def run_prompt_set_with_dependencies(set_name: str, prompt_sets: dict, document_text: str):
    if set_name not in prompt_sets:
        print(f" No prompt set found with name: {set_name}")
        return

    config = prompt_sets[set_name]
    system_prompt = config["system_prompt"]
    prompts = config["prompts"]

    print(f"\nğŸŸ© Running Agent: {set_name.upper()}\n" + "-" * 40)

    context = {}
    sub_prompts = {}

    # Separate sub-prompts from main prompts
    for key, instruction in prompts.items():
        if key == "sub-prompts":
            sub_prompts = instruction
        else:
            full_prompt = (
                f"Document:\n{document_text}\n\n"
                f"Task:\n{instruction}\n\n"
                "Respond directly and concisely. Only return the answer."
            )

            response = client.chat.completions.create(
                model="gpt-4o",
                messages=[
                    {"role": "system", "content": system_prompt},
                    {"role": "user", "content": full_prompt}
                ],
                temperature=0.1
            )

            answer = response.choices[0].message.content.strip()
            context[key] = answer
            print(f"{key}:\n{textwrap.fill(answer,width=80)}\n")

    # Now handle sub-prompts with context
    if sub_prompts:
        for sub_key, instruction in sub_prompts.items():
            formatted_input = (
                f"Document:\n{document_text}\n\n"
                f"Previously extracted:\n" +
                "\n".join([f"{k}: {v}" for k, v in context.items()]) +
                f"\n\nTask:\n{instruction}\n\n"
                "Respond directly and concisely. Only return the answer."
            )

            response = client.chat.completions.create(
                model="gpt-4o",
                messages=[
                    {"role": "system", "content": system_prompt},
                    {"role": "user", "content": formatted_input}
                ],
                temperature=0.1
            )

            sub_answer = response.choices[0].message.content.strip()
            print(f"{sub_key}:\n{sub_answer}\n")

# Optional: Run all sets (uses dependency-aware runner where needed)
def run_all_prompt_sets(prompt_sets: dict, document_text: str, only: str = None):
    if only:
        run_prompt_set_with_dependencies(only, prompt_sets, document_text)
    else:
        for set_name in prompt_sets:
            run_prompt_set_with_dependencies(set_name, prompt_sets, document_text)

# Run everything by default
run_all_prompt_sets(prompt_sets, document_text)

# Search or run a specific agent like this:
# run_all_prompt_sets(prompt_sets, document_text,
```
